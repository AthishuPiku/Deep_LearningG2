```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=100,
                    n_features=2,
                    n_informative=2,
                    n_redundant=0,
                    n_classes=2,
                    n_clusters_per_class=1,
                    class_sep=1,
                    random_state=42
                    )
```

```python
y.shape
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
# activation function
def step(z):
  return 1 if z>0 else 0

def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def log_Reg(x,y):
  x=np.insert(x,0,1,axis=1)
  weight= np.ones(x.shape[1])
  le=0.1

  for i in range(200):
    j=np.random.randint (0,100)
    y_hat=sigmoid(np.dot(x[j],weight))
    weight=weight+le*(y[j]-y_hat)*x[j]
  return weight[0],weight[1:]
```

```python
log_Reg(x,y)
```

```python
intercept_,coef_=log_Reg(x,y)
```

```python
m=-coef_[0]/coef_[1]
c=-intercept_/coef_[1]
```

```python
x_train=np.linspace(-3,3,100)
y_pred=m*x_train+c
```

```python
plt.plot(x_train, y_pred,color="red")
plt.scatter(x[:,0],x[:,1],c=y)
plt.ylim(-3,2)
plt.show()
```
