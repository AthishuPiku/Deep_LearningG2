```python
import opendatasets as od
od.download("https://www.kaggle.com/competitions/playground-series-s5e8")
```

```python
train="/content/playground-series-s5e8/train.csv"
```

```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv(train)
```

```python
train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
train.shape
```

```python
train.columns
```

```python
for x in train.columns:
  if train[x].dtype=='object':
   train[x]= LE.fit_transform(train[x])
```

```python
train.sample(2)
```

```python
test=pd.read_csv("/content/playground-series-s5e8/test.csv")
```

```python
for x in test.columns:
  if test[x].dtype=='object':
   test[x]= LE.fit_transform(test[x])
```

```python
test.sample()
```

```python
x_train=train.drop(["id","y"],axis=1)
y_train=train["y"]
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
LR.score(x_train,y_train)
```

```python
y_pred=LR.predict(test.drop("id",axis=1))
```
