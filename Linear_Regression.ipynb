### **Linear Regression**

Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to the observed data. It aims to find the "line of best fit" that minimizes the difference between the observed values and the values predicted by the model. This line is defined by a set of coefficients (slope and intercept) that quantify the relationship between the variables.

image.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples = 1000, n_features=1, n_informative=0,n_targets=1,random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
LR.score(x_test,y_test)
```

```python
x_train
```

```python
y_train
```

```python
LR.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print(mean_absolute_error(y_test,LR.predict(x_test)))
print(mean_squared_error(y_test,LR.predict(x_test)))
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,LR.predict(x_test),color='red')
plt.show()
```
